```python
#Histogram
```

```python
data3=pd.read_csv("SalaryGender.csv")
data3.head()
```

```python
plt.hist(data3["Salary"],np.sort(data3["Age"]),rwidth=.99)
plt.title("Histogram with Matplot",fontsize=30,color="r")
plt.show()
```

```python
plt.hist(data3["Salary"],rwidth=.5,color="aqua")
plt.show()
```

```python
sns.distplot(data3["Salary"],color="g",label="salary",)
plt.title("Histogram with seaborn",color="r",fontsize=30)
plt.xlabel("age",fontsize=20)
plt.ylabel("salary",fontsize=20)
plt.legend(loc=0)
plt.show()
```

```python
sns.distplot(data3["Salary"])
plt.show()
```
